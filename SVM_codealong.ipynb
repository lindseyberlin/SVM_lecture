```python
import pandas as pd
import numpy as np

from sklearn import svm, datasets, metrics

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# make toy data
X, y = datasets.make_blobs(n_samples=100,
                          n_features=2, 
                          centers=2,
                          cluster_std=2,
                          random_state=321)
```

```python
classifier = svm.SVC(kernel="linear")
classifier.fit(X, y)
y_pred = classifier.predict(X)
```

```python
metrics.confusion_matrix(y, y_pred)
```

```python
classifier.decision_function(X)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1], c=y, cmap="jet")

# Plot the SVM line between the clusters!
# get current axis
ax = plt.gca()

#get the xlim, ylim
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# creating a meshgrid
xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
XX, YY = np.meshgrid(xx, yy)

# creating the x-y pairs
xy = np.vstack([XX.ravel(), YY.ravel()]).T

Z = classifier.predict(xy).reshape(XX.shape)

ax.contour(XX, YY, Z)

plt.show()
```

```python
# An even less linear example

def twospirals(n_points, noise=.5):
    """
     Returns the two spirals dataset.
    """
    n = np.sqrt(np.random.rand(n_points,1)) * 780 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points,1) * noise
    d1y = np.sin(n)*n + np.random.rand(n_points,1) * noise
    return (np.vstack((np.hstack((d1x,d1y)),np.hstack((-d1x,-d1y)))), 
            np.hstack((np.zeros(n_points),np.ones(n_points))))

X2, y2 = twospirals(1000)
```

```python
classifier = svm.SVC(kernel="linear")
classifier.fit(X2, y2)
y_pred2 = classifier.predict(X2)
```

```python
metrics.confusion_matrix(y2, y_pred2)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X2[:,0], X2[:,1], c=y2, cmap="jet")

# Plot the SVM line between the clusters!
# get current axis
ax = plt.gca()

#get the xlim, ylim
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# creating a meshgrid
xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
XX, YY = np.meshgrid(xx, yy)

# creating the x-y pairs
xy = np.vstack([XX.ravel(), YY.ravel()]).T

Z = classifier.predict(xy).reshape(XX.shape)

ax.contour(XX, YY, Z)

plt.show()
```

```python
# Let's do not a straight separating line!
classifier2 = svm.SVC(gamma="auto")
classifier2.fit(X2, y2)
y_pred2 = classifier2.predict(X2)
```

```python
metrics.confusion_matrix(y2, y_pred2)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X2[:,0], X2[:,1], c=y2, cmap="jet")

# Plot the SVM line between the clusters!
# get current axis
ax = plt.gca()

#get the xlim, ylim
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# creating a meshgrid
xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
XX, YY = np.meshgrid(xx, yy)

# creating the x-y pairs
xy = np.vstack([XX.ravel(), YY.ravel()]).T

Z = classifier2.predict(xy).reshape(XX.shape)

ax.contour(XX, YY, Z)

plt.show()
```
